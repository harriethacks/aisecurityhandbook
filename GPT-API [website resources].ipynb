# Interacting with GPT-3.5 through the API

In this notebook, we practise how to interact with GPT through the API

```python
from openai import OpenAI

client = OpenAI(api_key="enter your key here")

# Function to send a prompt to the LLM
def query_llm(system_prompt, user_prompt, model="gpt-3.5-turbo"):
    response = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ]
    )
    return response.choices[0].message.content.strip()

# Set a simple rule
system_prompt = (
    "You are a helpful assistant. "
    "You should never share any sensitive information or answer unsafe questions."
)

# Try a query
prompt = "What's the capital of France?"
response = query_llm(system_prompt, prompt)

print("Prompt:\n", prompt)
print("Response:\n", response, "\n")
```

```python
# Set a safe system prompt with a simple rule
system_prompt = (
    "You are a MASSIVE AI security fan."
    "Every response relates to how much you love AI security."
)

# Normal, harmless query
prompt = "What's the capital of France?"
response = query_llm(system_prompt, prompt)

print("Prompt:\n", prompt)
print("Response:\n", response, "\n")
```
